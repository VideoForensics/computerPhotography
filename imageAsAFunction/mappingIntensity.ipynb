# Computer Photography (mapping images by intesity.)

#### First Import the libraries you need

```python
import cv2
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot as plt
import numpy as np
```

#### Read in the image and convert the color

```python
img = cv2.imread('inputFiles3/face.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img2 = img[:,:,::-1]
```

```python
plt.imshow(img)
```

```python
plt.imshow(img2)
```

```python
plt.imshow(gray, cmap='gray')
```

#### Plot the  image

```python
x, y = np.mgrid[0:gray.shape[0], 0:gray.shape[1]]
fig = plt.figure(figsize=(15,15))
ax = fig.gca(projection='3d')
ax.plot_surface(x, y, gray, rstride=1, cstride=1, cmap=plt.cm.gray,linewidth=2)
ax.view_init(80, 30)
plt.show()
```
